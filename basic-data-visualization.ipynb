```python
import numpy as np
import matplotlib.pyplot as plt

# Optional: use this in Jupyter Notebook for inline plots
# %matplotlib inline

# Function to add labels on bars
def addlabels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha='center')

# Corrected data
company = ['Google', 'Amzn', 'Msft']
revenue = [90, 136, 85]
profit = [40, 7, 34]
xpos = np.arange(len(company))

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(xpos, revenue, width=0.4, label='Revenue')
plt.bar(xpos - 0.4, profit, width=0.4, label='Profit')

# Add labels to the bars
addlabels(xpos, revenue)
addlabels(xpos - 0.4, profit)

# Set x-ticks and labels
plt.xticks(xpos, company)

# Add labels and title
plt.ylabel('Revenue (BLN)')
plt.title('US Technology Stocks')
plt.legend()

# Display the plot
plt.show()
```
